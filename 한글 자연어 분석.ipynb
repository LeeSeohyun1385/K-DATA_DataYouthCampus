```python
#불용어 제거
stop_words=[',','(',')',',',"'",'%','-','X',').','x',
            '의','자','에','안','번','호','을','이','다','만',
           '로','가','를','제','위','월','수','중','것','표',
           '멍','및','생','략','명']
ko = [each_word for each_word in ko if each_word not in stop_words] 
#ko에 있는 코드 중 stop_words조건을 만족하는 단어 제거
ko = nltk.Text(ko, name='국회법안')
```

```python
#연어분석 - 단어들이 서로 연결되어 사용되는 단어들 추출
from konlpy.tag import Kkma
from konlpy.corpus import kolaw
from konlpy.utils import pprint
from nltk import collocations

measures = collocations.BigramAssocMeasures()
#어느 범위까지 찾을것인가
tagged_words = Kkma().pos(doc_ko) #형태소분석결과 저장

stopwords = [u'에게'] #추가
words = [w for w, t in tagged_words] #추가


#연어찾기
finder = collocations.BigramCollocationFinder.from_words(words)
#tagged_words: {형태소 : 분사}  dict형태로 저장되어있음 
#형태소에 해당하는 부분만 words로 만듬 (저장)

# 연어찾는 단위 설정
finder.apply_word_filter(lambda w: len(w) < 2 or w in stopwords) # or w in stopwords 추가
finder.apply_freq_filter(4)
#연어길이가 2이하인것, 불용어 단어--> 제거
pprint(finder.nbest(measures.pmi,10)) #20
```

### 워드클라우드 
- WordCloud().generate_from_frequencies(data(임의).vocab())

```python
#워드클라우드 만들기
from wordcloud import WordCloud, STOPWORDS

#vocab에서 상위빈도 단어들 300개 가져와 출력
data = ko.vocab().most_common(300)

wordcloud = WordCloud(font_path = 'c:\Window\Fonts\malgun.ttf',
                      background_color='white',relative_scaling=0.2).generate_from_frequencies(dict(data))
#dict- 단어,빈도 이런식으로 속성이 나뉘어 있음
#relative_scaling :빈도 크기에 차이에 따른 크기??

plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
