```python
#중복
taxi_driver = taxi_driver.drop_duplicates(['ID','text'],ignore_index=True)
```

```python
taxi_driver
```

```python
#댓글이 비어있는 경우 제거 
#taxi_driver['text'].isnull().sum()
taxi_driver = taxi_driver.dropna(axis=0)
```

```python
taxi_driver
```

```python
#인덱스 재정렬
taxi_driver = taxi_driver.reset_index(drop=True)
taxi_driver
```

```python
# &#34; -> "
# &#39 -> '
# &gt; -> >
# &lt; -> <
```

```python
#taxi_driver = taxi_driver.str.replace('&#34;','"')
```

```python
index = taxi_driver['text'][taxi_driver['text'].str.contains('&gt;')==True].index
index

ind_list = list(index)
ind_list

for i in ind_list:
    taxi_driver['text'][i] = taxi_driver['text'][i].replace('&gt;',">")
    
```
