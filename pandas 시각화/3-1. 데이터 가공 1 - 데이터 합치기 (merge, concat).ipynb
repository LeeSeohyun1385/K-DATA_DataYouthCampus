# 4. 데이터 가공 1 - 데이터 통합
## 4-1. pandas.merge()

```python
#샘플 데이터 생성
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남'], ['신사임당',48,'여']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['홍길동','터미네이터',5],['홍길동','비긴어게인',2], 
                 ['홍길동','살인의추억',4], ['신사임당','비긴어게인',5],
                ['장보고','살인의추억',3]], 
                           columns = ['이름','영화','평점'])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
#df1.merge(df2)
```

```python
pd.merge(df1,df2,on='이름',how='inner')
#on: 동일한 컬럼 지정 (동일한 컬럼명이 2개 이상 있을 때  다 같은 경우에만 출력)
#동일한 컬럼명이 두개지만 하나만 합치고 싶을 때 등등
```

```python
df3 = DataFrame([['홍길동',23,'남'],['이순신',55,'남'], ['신사임당',48,'여']], 
                           columns = ['사용자명','나이','성별'])
df4 = DataFrame([['홍길동','터미네이터',5],['홍길동','비긴어게인',2], 
                 ['홍길동','살인의추억',4], ['신사임당','비긴어게인',5],
                ['장보고','살인의추억',3]], 
                           columns = ['이름','영화','평점'])
```

```python
df3
```

```python
df4
```

```python
#컬럼명 다른 경우( 같은 것이 하나도 없는 경우)
pd.merge(df3,df4,left_on='사용자명',right_on='이름',how='inner')
```

## 4-2. pandas.concat(axis = 0)
* 동일한 컬럼들을 가지고 있는 데이터프레임들을 하나로 합치고 싶은 경우

```python
#row방향으로 합침
#기준 : columns index
```

```python
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['신사임당',48,'여'], ['장보고',22,'남']], 
                           columns = ['이름','나이','성별'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],axis = 0)
#컬럼명이 같아야 잘합쳐짐
```

```python
pd.concat([df1,df2],axis = 0,ignore_index = True) #컬럼명 다를 경우 
```

## 4-3. concat(axis = 1)
* 서로 다른 컬럼을 가진 데이터프레임들을 로우 인덱스의 값을 기준으로 합치고 싶은 경우

```python
#column방향으로 합침
#기준 : row index
```

```python
# 샘플 데이터
df1 = DataFrame([[23,'남'],[55,'남'], [48,'여']], 
                           columns = ['나이','성별'], index = ['홍길동','이순신','신사임당'])
df2 = DataFrame([['서울',3], ['경기', 1]], 
                           columns = ['주소','학년'], index = ['홍길동','신사임당'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],axis = 1)
```

```python
#merge: 기준 column이 있어 같은 컬럼에 있는 값끼리 (행방향으로)각각 비교해서 합침
#concat: 옆에 그냥 이어 붙임 ( 비교하거나 하는 수행은 안함 : row인덱스가 같은 애들끼리 이어붙이는 비교는 수행)
#merge와 다르게 concat 3개 이상을 한번에 수행할 수 있음 [list기호 활용]
#concat: low index가 unique해야함
#inner 두데이터에 모두 있는 경우에만 출력
#outer 둘다 없어도 left,right 설정해서 한데이터는 다 붙여주는것??? 
```

```python
#NaN 포함된 이순신 삭제
pd.concat([df1,df2],axis = 1,join='inner')
```
