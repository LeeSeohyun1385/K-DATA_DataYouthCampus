# 한국어 자연어 처리(NLP)_KoNLPy 
## 토큰화(Tokenization) - 형태소 토큰화
### 형태소 분석기 - 꼬꼬마(Kkma), 한나눔(Hannanum), Okt(Open Korea Text), 메캅(Mecab), 코모란(Komoran)

### Kkma()

```python
from konlpy.tag import Kkma
```

```python
kkma = Kkma()

kkma.sentences('한국어 분석을 시작합니다. 재미있어요.~~~~~')
#텍스트를 문장 단위로 끊음
```

```python
# 명사
kkma.nouns('한국어 분석을 시작합니다. 재미있어요.~~~~~')
```

```python
# 형태소 분석(품사태깅)
kkma.pos('한국어 을분석 시작합니다. 재미있어요.~~~~~')
```

### Hannanum()

```python
from konlpy.tag import Hannanum
```

```python
hannanum=Hannanum()

# 명사 추출
hannanum.nouns('한국어 분석을 시작합니다. 재미있어요.~~~~~')
```

```python
# 형태소 분석
hannanum.morphs('한국어 분석을 시작합니다. 재미있어요.~~~~~')
```

```python
hannanum.pos('한국어 분석을 시작합니다. 재미있어요.~~~~~')
```

### Okt()
-  morphs : 형태소 추출
-  pos : 품사 태깅(Part-of-speech tagging)
-  nouns : 명사 추출

```python
from konlpy.tag import Okt
```

```python
t=Okt()

t.nouns('한국어 분석을 시작합니다. 재미있어요.~~~~~')
```

```python
t.morphs('한국어 분석을 시작합니다. 재미있어요.~~~~~')
```

```python
t.pos('한국어 분석을 시작합니다. 재미있어요.~~~~~')
```

# 워드클라우드

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
text = open(r'C:\Users\Lee seohyun\Downloads\09. alice.txt').read()
```

```python
text
```

### 영어 불용어(Stop word) 제거

```python
#stopwords불러오기
stopwords = set(STOPWORDS)
```

```python
stopwords
```

### wordcloud 생성
- font_path	: 한글의 경우 깨지지 않게 폰트 경로 지정
- background_color : 백그라운드 색 지정
- stopwords : 불용어 지정
- mask :이미지에 워드클라우드 출력

```python
wc = WordCloud(background_color='white', max_words=2000, stopwords=stopwords)
wc = wc.generate(text)  # 데이터 불러와서 워드클라우드 옵션 지정
wc
```

- 빈도수

```python
# 주요 키워드 순 / 상대빈도 순
wc.words_
```

### 워드클라우드 이미지 출력

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,12))
plt.imshow(wc, interpolation='bilinear') #interpolation : 이미지 부드럽기 정도
plt.axis('on')   #축이 있는 경우 , off : 축 숫자 제거 
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')   #축이 없는 경우
plt.show()
```

### 추가적으로 불용어 제거

```python
stopwords = set(STOPWORDS)
stopwords.add('said')
stopwords.add('say')
stopwords.add('know')
stopwords.add('one')
wc = WordCloud(background_color='white', max_words=2000, stopwords=stopwords)
wc = wc.generate(text)  # 데이터 불러와서 워드클라우드 옵션 지정
wc
```

```python
plt.figure(figsize=(16,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')   #축이 없는 경우
plt.show()
```
